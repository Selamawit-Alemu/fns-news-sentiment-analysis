```python

# Step 0: Setup - Import libraries and add project root to sys.path
import sys
import os
sys.path.append(os.path.abspath('..'))

import matplotlib.pyplot as plt
import pandas as pd

# Import your custom modules
from src.data_loader import load_all_stock_data
from src.indicators import add_indicators

# Step 1: Load all stock data from yfinance_data folder
folder_path = '../data/yfinance_data'
all_stock_data = load_all_stock_data(folder_path=folder_path)

print(f"Loaded data for stocks: {list(all_stock_data.keys())}")

# Step 2: Add technical indicators and visualize for each stock
for ticker, df in all_stock_data.items():
    print(f"\nProcessing {ticker}...")

    # Add SMA, RSI, MACD indicators
    df_with_indicators = add_indicators(df)

    # Display head of dataframe with indicators
    display(df_with_indicators.head())

    # Plot Close Price and SMA 20
    plt.figure(figsize=(14, 7))
    plt.title(f"{ticker} Close Price and SMA (20)")
    plt.plot(df_with_indicators.index, df_with_indicators['Close'], label='Close Price')
    plt.plot(df_with_indicators.index, df_with_indicators['SMA_20'], label='SMA 20')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.show()

    # Plot RSI with threshold lines (overbought=70, oversold=30)
    plt.figure(figsize=(14, 4))
    plt.title(f"{ticker} RSI (14)")
    plt.plot(df_with_indicators.index, df_with_indicators['RSI_14'], label='RSI 14')
    plt.axhline(70, color='red', linestyle='--', label='Overbought (70)')
    plt.axhline(30, color='green', linestyle='--', label='Oversold (30)')
    plt.xlabel('Date')
    plt.ylabel('RSI')
    plt.legend()
    plt.show()

    # Plot MACD and MACD Signal line
    plt.figure(figsize=(14, 4))
    plt.title(f"{ticker} MACD")
    plt.plot(df_with_indicators.index, df_with_indicators['MACD'], label='MACD')
    plt.plot(df_with_indicators.index, df_with_indicators['MACD_signal'], label='Signal Line')
    plt.xlabel('Date')
    plt.ylabel('MACD')
    plt.legend()
    plt.show()

```

```python
import pynance as pn
import talib
import matplotlib.pyplot as plt

# Fetch data for AAPL using the correct pynance function
df = pn.data.get('AAPL')

# Calculate SMA 20 with TA-Lib
df['SMA_20'] = talib.SMA(df['Close'], timeperiod=20)

# Calculate daily returns manually
df['daily_returns'] = df['Close'].pct_change()

# Calculate volatility (std dev of returns) and cumulative return
volatility = df['daily_returns'].std()
cumulative_return = (1 + df['daily_returns']).prod() - 1

print("Volatility:", volatility)
print("Cumulative Return:", cumulative_return)

# Plot closing prices and SMA
plt.figure(figsize=(12,6))
plt.plot(df['Close'], label='Close')
plt.plot(df['SMA_20'], label='SMA 20')
plt.legend()
plt.show()
```
